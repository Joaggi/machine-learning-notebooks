```
def suma(operando_1, operando_2):
  if (operando_1 > 10):
    raise Exception("Error")
  return operando_1 + operando_2
```

```
def multiplicacion(operando_1, operando_2):
  return operando_1 * operando_2
```

```
def division(operando_1, operando_2):
  return operando_1 / operando_2
```

```
def operacion(operando_1, operando_2, operando_3, operando_4):
  operacion_1 = suma(operando_1, operando_2)
  operacion_2 = multiplicacion(operacion_1, operando_3)
  operacion_3 = division(operacion_2, operando_4)
  return operacion_3
```
