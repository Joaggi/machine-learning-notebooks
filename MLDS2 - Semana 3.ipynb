```python
iris = load_iris()
```

```python
logistic = LogisticRegression(solver='saga', tol=1e-2, max_iter=200,
                              random_state=0)
```

```python
distributions = dict(C=uniform(loc=0, scale=1000000),
                     penalty=['l2', 'l1'])
```

```python
RandomizedSearchCV(n_iter=50)
```

```python
clf = RandomizedSearchCV(logistic, distributions, random_state=0)
```

```python
search = clf.fit(iris.data, iris.target)
search.best_params_
```

```python
import pandas as pd
pd.DataFrame(search.cv_results_).sort_values("mean_test_score", ascending=False)
```

# Redes neuronales

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X, y = make_classification(n_samples=100, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                     random_state=1)

X.shape
y
```

Por defecto, Sklearn usa 1 sola capa oculta y 100 neuronas.

[Redes Neuronales](http://playground.tensorflow.org/#activation=relu&regularization=L2&batchSize=10&dataset=spiral&regDataset=reg-plane&learningRate=0.03&regularizationRate=0.1&noise=0&networkShape=7,8,8&seed=0.33146&showTestData=false&discretize=false&percTrainData=50&x=true&y=true&xTimesY=false&xSquared=false&ySquared=false&cosX=false&sinX=false&cosY=false&sinY=false&collectStats=false&problem=classification&initZero=false&hideText=false)

```python
clf = MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)
clf.predict_proba(X_test[:1])
clf.predict(X_test[:5, :])
clf.score(X_test, y_test)
```

Cuando deseo más capas y más neuronas utilizo `hidden_layer_sizes`. Cada número representa el número de neuronas ocultas de la capa dada.

```python
clf = MLPClassifier(hidden_layer_sizes=(100, 100, 100, 100), random_state=1, max_iter=300).fit(X_train, y_train)
clf.predict_proba(X_test[:1])
clf.predict(X_test[:5, :])
clf.score(X_test, y_test)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100, 200, 100, 20, 20), random_state=1, max_iter=300).fit(X_train, y_train)
clf.predict_proba(X_test[:1])
clf.predict(X_test[:5, :])
clf.score(X_test, y_test)
```

# Funciones dentro de funciones

```python
import numpy as np
#Función para generar la función de predicción de un
# clasificador entrenado previamente.
def gen_pred_fun(clf):
    def pred_fun(x1, x2):
        x = np.array([[x1, x2]])
        return clf.predict(x)
    return pred_fun
```

```python
class Model():
  def predict(self, x):
    return x
```

```python
clf_model = Model()
clf_model.predict(2)
```

```python
fun = gen_pred_fun(clf_model)
```

```python
fun(2, 3)
```
